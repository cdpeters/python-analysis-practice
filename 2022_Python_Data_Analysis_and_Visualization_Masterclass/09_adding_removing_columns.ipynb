### Finding the Highest price/sqft Homes

```python
houses_df["price_sqft"] = houses_df["price"] / houses_df["sqft_living"]

houses_df
```

```python
houses_df.sort_values("price_sqft", ascending=False)
```

```python
houses_df.sort_values("price_sqft", ascending=False).head(50)["zipcode"].value_counts().plot(kind="bar")
```

### Finding Largest Bitcoin Price Changes

```python
btc_df.set_index("Date", inplace=True)
```

```python
btc_df["Change"] = btc_df["Close"] - btc_df["Open"]
```

```python
btc_df.sort_values("Change", ascending=False)
```

```python
btc_df["Delta"] = btc_df["High"] - btc_df["Low"]
```

```python
btc_df.sort_values("Delta", ascending=False).head(10)["Delta"].plot(kind="bar")
```

```python
btc_df.sort_values("Delta", ascending=False).head(10)["Delta"].plot(kind="bar")
```
