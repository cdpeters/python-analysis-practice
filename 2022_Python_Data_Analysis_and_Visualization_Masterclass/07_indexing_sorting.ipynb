# Section 7: Indexing & Sorting

```python
from pathlib import Path

import pandas as pd
```

```python
# Path to data directory
data_dir = Path.cwd() / "course_materials" / "data"
```

```python
# King's County housing dataset
kc_houses_df = pd.read_csv(data_dir / "kc_house_data.csv")

# Titanic Passenger dataset
titanic_df = pd.read_csv(data_dir / "titanic.csv")

# Netflix dataset
netflix_df = pd.read_csv(data_dir / "netflix_titles.csv", sep="|", index_col=0)

# Bitcoin dataset
btc_df = pd.read_csv(data_dir / "coin_Bitcoin.csv")

# World Happiness Index dataset
world_happiness_df = pd.read_csv(data_dir / "world-happiness-report-2021.csv")
```

### Set_Index Basics

### Setting index with read_csv

```python
# World Happiness Index dataset
new_world_happiness_df = pd.read_csv(data_dir / "world-happiness-report-2021.csv", index_col="Country name")
new_world_happiness_df.head()
```

### sort_values()

```python
world_happiness_df.sort_values(by="Healthy life expectancy", ascending=False, inplace=True)
```

### sorting by multiple columns

```python
kc_houses_df.sort_values(by=["bedrooms", "bathrooms"], ascending=False)
```

### sorting text columns

```python
titanic_df.sort_values(by="name")
```

```python
# `name` in the lambda function here is a Series, thus `str` is used so that the
# supplied `lower()` function is vectorized (applied element-wise)
titanic_df.sort_values(by="name", key=lambda name: name.str.lower())
```

### sort_index

```python
world_happiness_df.sort_index()
```

### Sorting and Plotting

```python
titanic_df["pclass"].value_counts()
```

```python
titanic_df["pclass"].value_counts().plot(kind="bar")
```

```python
# Sorting by the index
titanic_df["pclass"].value_counts().sort_index().plot(kind="bar")
```

```python
titanic_df["pclass"].value_counts().sort_values()
```

```python
# Sorting by the values in ascending order
titanic_df["pclass"].value_counts().sort_values().plot(kind="bar")
```

```python
kc_houses_df["bedrooms"].value_counts()
```

```python
kc_houses_df["bedrooms"].value_counts().sort_values().plot(kind="bar")
```

```python
kc_houses_df["bedrooms"].value_counts().sort_index().plot(kind="bar")
```

### loc

```python
# Returning a Series
world_happiness_df.loc["Zambia"]
```

```python
# Returning a DataFrame
world_happiness_df.loc[["Zambia"]]
```

```python
world_happiness_df.loc[["Canada", "Mexico", "United States"]]
```

```python
# `loc[start:stop:step]`. Note that the `stop` is included with `loc` slicing
titanic_df.loc[2:8:2]
```

```python
world_happiness_df.sort_index(inplace=True)
```

```python
# Can even provide a step here if needed
world_happiness_df.loc["Denmark":"Egypt"]
```

### iloc

```python
# `iloc[start:stop:step] where `stop` is not included
world_happiness_df.iloc[0:3]
```

```python
kc_houses_df.sort_index(ascending=False, inplace=True)
```

```python
kc_houses_df
```

```python
kc_houses_df.iloc[0:3]
```

### loc and iloc with Series

```python
world_happiness_df.loc["Denmark":"Egypt", ["Healthy life expectancy", "Ladder score"]]
```

```python
# Slower approach
world_happiness_df.loc["Denmark":"Egypt"][["Healthy life expectancy", "Ladder score"]]
```

```python
world_happiness_df.loc["Denmark":"Egypt", "Ladder score":"Healthy life expectancy"]
```

```python
# On a Series
kc_houses_df["bedrooms"].value_counts().iloc[0:3]
```

```python
# looks for the label 18, here the label is not a number, it is a string so
# `str()` or `"18"` need to be used as the argument to `loc`
titanic_df["age"].value_counts().loc[str(18)]
```
