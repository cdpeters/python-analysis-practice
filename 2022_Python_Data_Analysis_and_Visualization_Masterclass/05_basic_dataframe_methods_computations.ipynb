```python
from pathlib import Path

import pandas as pd
```

```python
pd.__version__
```

```python
# Path to data directory
data_dir = Path.cwd() / "course_materials" / "data"
```

#### Min & Max

```python
# King's County housing dataset
kc_houses_df = pd.read_csv(data_dir / "kc_house_data.csv")

kc_houses_df.head()
```

```python
kc_houses_df.min()
```

```python
type(kc_houses_df.min())
```

#### Sum & Count

```python
kc_houses_df.sum(numeric_only=True)
```

```python
# Titanic Passenger dataset
titanic_df = pd.read_csv(data_dir / "titanic.csv")

titanic_df.head()
```

```python
titanic_df.sum()
```

```python
# Netflix dataset
netflix_df = pd.read_csv(data_dir / "netflix_titles.csv", sep="|", index_col=0)

netflix_df.head()
```

```python
netflix_df.count()
```

#### Mean, Median, & Mode

```python
kc_houses_df.mean(numeric_only=True)
```

```python
titanic_df.mean(numeric_only=True)
```

```python
titanic_df.mode(numeric_only=True)
```
