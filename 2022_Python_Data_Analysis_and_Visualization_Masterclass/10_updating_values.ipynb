# Section 10: Updating Values

```python
from pathlib import Path

import pandas as pd
```

```python
# Path to data directory
data_dir = Path.cwd() / "course_materials" / "data"
```

```python
####### NOTE DATAFRAME NAME CHANGES TO MATCH COURSE
# King's County housing dataset
houses_df = pd.read_csv(data_dir / "kc_house_data.csv")

# Titanic Passenger dataset
titanic_df = pd.read_csv(data_dir / "titanic.csv")

# Netflix dataset
netflix_df = pd.read_csv(data_dir / "netflix_titles.csv", sep="|", index_col=0)

# Bitcoin dataset
btc_df = pd.read_csv(data_dir / "coin_Bitcoin.csv")

# World Happiness Index dataset
countries_df = pd.read_csv(
    data_dir / "world-happiness-report-2021.csv", index_col="Country name"
)
```

### The replace() method

```python
titanic_df["sex"].replace("female", "F")
```

```python
# replacing multiple values at once
titanic_df["sex"].replace(["female", "male"], ["F", "M"], inplace=True)
titanic_df
```

```python
titanic_df["age"].replace(["?"], [None], inplace=True)
titanic_df
```

```python
titanic_df["age"].value_counts(dropna=False)
```

### Updating Values Using loc[]

```python
countries_df
```

```python
countries_df.loc["Brazil", ["ladder_score", "Healthy life expectancy"]]
```

```python
countries_df.loc[["Brazil", "Panama", "Honduras"], ["ladder_score", "Healthy life expectancy"]]
```

```python
# Now here is how we use the notation above to update values
countries_df.loc[["Denmark", "Sweden", "Norway"], ["regional_indicator"]] = "Scandanavia"
```

```python
countries_df.loc[["Denmark", "Sweden", "Norway"]]
```

```python
# A silly example
countries_df.loc["Finland", ["upperwhisker", "lowerwhisker"]] = "MEOW"
countries_df
```

### Updating Multiple Values Using loc[]

```python
countries_df.loc[["Finland", "Denmark", "Iceland", "Sweden", "Norway"]]
```

```python
countries_df.loc[["Finland", "Denmark", "Iceland", "Sweden", "Norway"],["regional_indicator", "is_cold"]] = ["Nordic Countries", "yes"]
```

```python
countries_df
```

### Making Updates With loc[] and Boolean Masks

```python
many_bedrooms = houses_df["bedrooms"] >= 10

houses_df.loc[many_bedrooms, ["bedrooms"]] = "TOO MANY!"
```

```python
houses_df[many_bedrooms]
```

```python
high_quality = houses_df["grade"] > 12
great_views = houses_df["view"] == 4
houses_df[high_quality & great_views]
```

```python
houses_df["luxury_level"] = "normal"
```

```python
houses_df.loc[high_quality & great_views, ["luxury_level"]] = "ultra"
```

```python
houses_df["luxury_level"].value_counts()
```
