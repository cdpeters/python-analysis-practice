# Section 8: Filtering DataFrames

```python
from pathlib import Path

import pandas as pd
```

```python
# Path to data directory
data_dir = Path.cwd() / "course_materials" / "data"
```

```python
# King's County housing dataset
kc_houses_df = pd.read_csv(data_dir / "kc_house_data.csv")

# Titanic Passenger dataset
titanic_df = pd.read_csv(data_dir / "titanic.csv")

# Netflix dataset
netflix_df = pd.read_csv(data_dir / "netflix_titles.csv", sep="|", index_col=0)

# Bitcoin dataset
btc_df = pd.read_csv(data_dir / "coin_Bitcoin.csv")

# World Happiness Index dataset
world_happiness_df = pd.read_csv(data_dir / "world-happiness-report-2021.csv")
```

### Filtering DataFrames With A Boolean Series

```python
titanic_short_df = titanic_df.head()

titanic_short_df
```

```python
titanic_short_df["sex"] == "female"
```

```python
titanic_short_df[titanic_short_df["sex"] == "female"]
```

### Filtering With Comparison Operators

```python
titanic_df[titanic_df["survived"] == 1]
```

```python
titanic_df[titanic_df["pclass"] != 1]
```

```python
kc_houses_df[kc_houses_df["price"] > 5000000]
```

### The Between Method

```python
kc_houses_df["bedrooms"].between(5, 7).value_counts()
```

```python
kc_houses_df[kc_houses_df["bedrooms"].between(5, 7)]
```

```python
kc_houses_df[kc_houses_df["grade"].between(11,13)]
```

### The isin() Method

```python
countries = ["India", "Japan", "South Korea"]
netflix_df["country"].isin(countries)
```

```python
netflix_df[netflix_df["country"].isin(countries)]
```

### Combining Conditions Using AND (&)

```python
women = titanic_df["sex"] == "female"
died = titanic_df["survived"] == 0

titanic_df[women & died]
```

```python
kc_houses_df[(kc_houses_df["waterfront"] == 1) & (kc_houses_df["price"] < 500000)]
```

### Combining Conditions Using AND (&)

```python
kc_houses_df[(kc_houses_df["yr_built"] >= 2014) | (kc_houses_df["yr_renovated"] >= 2014)]
```

```python
fincher = netflix_df["director"] == "David Fincher"
scoresese = netflix_df["director"] == "Martin Scorsese"
netflix_df[fincher | scoresese]
```

```python
# Another way to do the above with isin()
netflix_df[netflix_df["director"].isin(["David Fincher", "Martin Scoresese"])]
```

```python
netflix_df[(fincher | scoresese) & (netflix_df["release_year"] >= 2015)]
```

### Bitwise Negation

```python
titanic_short_df
```

```python
women = titanic_short_df["sex"] == "female"
titanic_short_df[~women]
```

### isna() and notna() Methods

```python
netflix_df.info()
```

```python
# Entries with missing directors
netflix_df[netflix_df["director"].isna()]
```

```python
# Entries with missing directors and missing cast
netflix_df[netflix_df["director"].isna() & netflix_df["cast"].isna()]
```

```python
# Entries with a director
netflix_df[netflix_df["director"].notna()]
```

### Filtering + Plotting Examples

```python
titanic_df[titanic_df["sex"] == "female"]["survived"].value_counts()
```

```python
titanic_df[titanic_df["sex"] == "male"]["survived"].value_counts()
```

```python
women = titanic_df["sex"] == "female"
```

```python
# plotting survival of females
titanic_df[women]["survived"].value_counts().plot(kind="pie")
```

```python
# plotting survival of males
titanic_df[~women]["survived"].value_counts().plot(kind="pie")
```

```python
kc_houses_df[kc_houses_df["price"] > 3000000]["zipcode"].value_counts().plot(kind="bar")
```
